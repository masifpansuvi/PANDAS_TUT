##### Resampling use instead of groupby

```python
data = {
    'Date': df,
    "Temp" : range(200,290)
}
```

```python
temp_data = pd.DataFrame(data)
```

```python
temp_data
```

```python
# TO APPLY RESAMPLE WE NEED TO MAKE DATE TO INDEX OF THE TABLE
```

```python
temp_data.set_index('Date', inplace=True)
```

```python
temp_data
```

```python
temp_data.resample('M').max()
```

```python
temp_data.resample('M').count()
```

```python
temp_data.resample('M').max().sum()
```
