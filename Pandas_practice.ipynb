##  ***3-->Resampling & Aggregating Data***

```python
# LOAD DATASET

electric_production = pd.read_csv('/content/Electric_Production.csv')
```

```python
# CHANGE DATE INTO DATE-TIME

electric_production['DATE'] = pd.to_datetime(electric_production['DATE'])
```

```python
# MAKE SOME NEW COLUMNS

electric_production['Year'] = electric_production['DATE'].dt.year
electric_production['Month'] = electric_production['DATE'].dt.month
electric_production['Day'] = electric_production['DATE'].dt.day
```

```python
electric_production
```

```python
# LET SET DATE TO AN INDEX

electric_production.set_index('DATE', inplace=True)
```

```python
electric_production
```

##### *Resample the data to monthly averages using .resample('M').mean()*

```python
# RESAMPLING

monthly_avg = electric_production.resample('M').mean()
```

```python
monthly_avg
```

##### *Resample the data to weekly sums using .resample('W').sum().*

```python
electric_production.resample('W').sum()
```

##### *Resample the data to monthly averages using .resample('M').mean()*

```python
electric_production.resample('Y').mean()
```

```python
electric_production.resample('Y').sum()
```

```python
electric_production.resample('Y').max()
```

### ***4--> Shifting Data for Trend Analysis***

###### *Create a new column "Previous Month Production" by shifting the data backward (.shift(1)).*

```python
electric_production['Previous_Month_Production'] = electric_production['Month'].shift(1)
```

```python
electric_production
```

##### *Create a new column "Next Month Projection" by shifting the data forward (.shift(-1)).*

```python
electric_production['Next_Monthly_Projection'] = electric_production['Month'].shift(-1)
```
